```python
#loops for season and episode codes

season_episode_list =[]

for season in range(1,4):
    if season != 3:
        for episode in range(1,21):
            ep_num = str(episode)
            if len(ep_num) <= 1:
                ep_num = str(0)+ep_num
            else: 
                ep_num
            se_ep = str(season) + ep_num
            season_episode_list.append(se_ep)
    else:
        for episode in range(1,22):
            ep_num = str(episode)
            if len(ep_num) <= 1:
                ep_num = str(0)+ep_num
            else: 
                ep_num
            se_ep = str(season) + ep_num
            season_episode_list.append(se_ep)
```
